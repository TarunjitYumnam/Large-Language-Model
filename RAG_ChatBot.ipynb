```python
import faiss
from langchain.docstore.document import Document

knowledge_base = ["Hello, I am a Chat Bot!",
                  "Yumnam_Jr Here, How can I help you Today!",
                  "Mid Semester exam will start from 15/02/2025",
                  "Students have to clear all the due"]
documents = [Document(page_content= sentence) for sentence in knowledge_base]
documents
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(documents,embedding_model)
```

```python
retriever = vectorstore.as_retriever(search_kwargs = {"k":1})
retriever
```

```python
def generate_content(query):
  response = retriever.get_relevant_documents(query)
  if response:
    return response[0].page_content
  else:
    return "Could not find an Asn!"
```

```python
generate_content("What is ")
```

```python
import faiss
from langchain.docstore.document import Document

knowledge_base = ["I am a Chat Bot",
                  "My name is Yumnam_Jr",
                  "The mid semester exam will start from 15-02-2025",
                  "Students must clear dues before exam"]
documents = [Document(page_content= sentence) for sentence in knowledge_base]
documents
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(documents,embedding_model)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k":1})
retriever
```

```python
def get_ans(query):
  answer = retriever.get_relevant_documents(query)
  if answer:
    return answer[0].page_content
  else:
    return "Could not get an Answer"
```

```python
get_ans("Any information for students?")
```
