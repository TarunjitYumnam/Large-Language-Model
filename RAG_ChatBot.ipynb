```python
import faiss
from langchain.docstore.document import Document
passages = [
    "I am a Chat Bot.",
    "Yumnam_Jr here. How can I help you?",
    "Mid Semester Examination will start from 15-02-2025.",
    "Regular Classes will be conducted from 03:20 PM on all working days as per regular timetable except during examination.",
    "There will be no class from 11:00 AM to 12:50 PM.",
    "Students are allowed in the examination after clearing all dues."
]
documents = [Document(page_content=passage) for passage in passages]
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
embedding_model = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(documents,embedding_model)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k":1})
```

```python
def get_retrieved_context(query):
  docs = retriever.get_relevant_documents(query)
  if docs:
    return docs[0].page_content
  return "I don't have information about that."
```

```python
while True:
  query = input("You: ")
  if query.lower() == "exit":
    break
  response = get_retrieved_context(query)
  print(f"Chatbot: {response}")
```
